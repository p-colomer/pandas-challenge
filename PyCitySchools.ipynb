```python
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
school_data.head()
```

```python
student_data.head()
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

# District Summary

```python
#Calculate the total number of schools
total_schools = school_data_complete['school_name'].nunique()
total_schools
```

```python
#Calculate the total number of students
total_students = school_data_complete['student_name'].count()
total_students
```

```python
#Calculate the total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
#Calculate the average math score
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
#Calculate the average reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
s = school_data_complete['math_score']
x = 70
maths_scores_pasing = s[s >= x]
#maths_scores_pasing
```

```python
students_pass_maths = maths_scores_pasing.count()
students_pass_maths
```

```python
percentage_passing_maths = students_pass_maths/total_students*100
percentage_passing_maths
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
s = school_data_complete['reading_score']
x = 70
reading_scores_pasing = s[s >= x]
#reading_scores_pasing
```

```python
students_pass_reading = reading_scores_pasing.count()
students_pass_reading
```

```python
percentage_passing_reading = students_pass_reading/total_students*100
percentage_passing_reading
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing
filterinfDataframe = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70) ]

#filterinfDataframe
```

```python
students_pass_math_reading = filterinfDataframe['student_name'].count()
students_pass_math_reading
```

```python
percentage_passing_both = students_pass_math_reading/total_students*100
percentage_passing_both
```

```python
#Create a dataframe to hold the above results
district_summary = {'Total Scools':[total_schools],
                       'Total Students':[total_students],
                       'Total Budget':[total_budget],
                       'Average Math Score':[average_math_score],
                       'Average Reading Score':[average_reading_score],
                       '% Passing Math':[percentage_passing_maths],
                       '% Passing Reading':[percentage_passing_reading],
                       '% Overall Passing':[percentage_passing_both]
    
}



district_summary_df = pd.DataFrame(district_summary)

#formatting
district_summary_copy = district_summary_df.copy()
district_summary_copy['Total Budget'] = district_summary_copy['Total Budget'].astype(float).map("${:,.2f}".format)
district_summary_copy
```

# School Summary

```python
group_school = school_data_complete.groupby(['school_name'])

average_reading_score = group_school['reading_score'].mean()
#average_reading_score 
```

```python
average_maths_score = group_school['math_score'].mean()
#average_maths_score
```

```python
size_schools = group_school.size()
#size_schools
```

```python
total_budget_school = group_school['budget'].mean()
#total_budget_school
```

```python
per_student_budget = total_budget_school/size_schools
#per_student_budget
```

```python
passed_maths_school = school_data_complete.loc[school_data_complete['math_score'] >= 70].groupby(['school_name']).size()
percentage_maths_schools = passed_maths_school/size_schools*100
#percentage_maths_schools
```

```python
passed_reading_school = school_data_complete.loc[school_data_complete['reading_score'] >= 70].groupby(['school_name']).size()
percentage_reading_schools = passed_reading_school/size_schools*100
#percentage_reading_schools
```

```python
passed_overall_school = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete["reading_score"]>=70)].groupby(['school_name']).size()
percentage_overall_school = passed_overall_school/size_schools*100
#percentage_overall_school
```

```python

school_type = school_data.set_index(["school_name"])["type"]
#school_type
```

```python
school_summary_df = pd.DataFrame({
    'School Type':school_type,
    'Total Students':size_schools,
    'Total School Budget':total_budget_school,
    'Per Student Budget':per_student_budget,
    'Average Math Score':average_maths_score,
    'Average Reading Score':average_reading_score,
    '% Passing Math':percentage_maths_schools,
    '% Passing Reading':percentage_reading_schools,
    '% Overall Passing':percentage_overall_school
})

#formatting
school_summary_df_copy = school_summary_df.copy()
school_summary_df_copy['Total School Budget'] = school_summary_df_copy['Total School Budget'].astype(float).map("${:,.2f}".format)
school_summary_df_copy['Per Student Budget'] = school_summary_df_copy['Per Student Budget'].astype(float).map("${:,.2f}".format)
school_summary_df_copy
```

# Top Performing Schools (By % Overall Passing)

```python
sorted_top_overall_passing_df = school_summary_df.sort_values(by = ['% Overall Passing'], ascending = False)

#formatting
sorted_top_overall_passing_df['Total School Budget'] = sorted_top_overall_passing_df['Total School Budget'].astype(float).map("${:,.2f}".format)
sorted_top_overall_passing_df['Per Student Budget'] = sorted_top_overall_passing_df['Per Student Budget'].astype(float).map("${:,.2f}".format)

sorted_top_overall_passing_df.head()
```

# Bottom Performing Schools (By % Overall Passing)

```python
sorted_bottom_overall_passing_df = school_summary_df.sort_values(by = ['% Overall Passing'], ascending = True)

#formatting
sorted_bottom_overall_passing_df['Total School Budget'] = sorted_bottom_overall_passing_df['Total School Budget'].astype(float).map("${:,.2f}".format)
sorted_bottom_overall_passing_df['Per Student Budget'] = sorted_bottom_overall_passing_df['Per Student Budget'].astype(float).map("${:,.2f}".format)

sorted_bottom_overall_passing_df.head()
```

# Math Scores by Grade

```python
columns_maths_score = ['school_name','grade','math_score']
group_grade = school_data_complete[columns_maths_score]    
group_grade =pd.DataFrame(group_grade)

grade_9th = group_grade.loc[(group_grade['grade'] == '9th')].groupby(['school_name']).mean()
grade_9th = grade_9th.rename(columns={"math_score":"9th"})

grade_10th = group_grade.loc[(group_grade['grade'] == '10th')].groupby(['school_name']).mean()
grade_10th = grade_10th.rename(columns={"math_score":"10th"})

grade_11th = group_grade.loc[(group_grade['grade'] == '11th')].groupby(['school_name']).mean()
grade_11th = grade_11th.rename(columns={"math_score":"11th"})

grade_12th = group_grade.loc[(group_grade['grade'] == '12th')].groupby(['school_name']).mean()
grade_12th = grade_12th.rename(columns={"math_score":"12th"})
```

```python
grades_9_10 = pd.merge(grade_9th, grade_10th, how="left", on=["school_name"])
grades_9_10_11 = pd.merge(grades_9_10, grade_11th, how="left", on=["school_name"])    
math_score_by_grade = pd.merge(grades_9_10_11, grade_12th, how="left", on=["school_name"])
math_score_by_grade
```

# Reading Score by grade

```python
columns_reading_score = ['school_name','grade','reading_score']
group_grade_reading = school_data_complete[columns_reading_score]    
group_grade_reading =pd.DataFrame(group_grade_reading)

grade_9th_reading = group_grade_reading.loc[(group_grade_reading['grade'] == '9th')].groupby(['school_name']).mean()
grade_9th_reading = grade_9th_reading.rename(columns={"reading_score":"9th"})

grade_10th_reading = group_grade_reading.loc[(group_grade_reading['grade'] == '10th')].groupby(['school_name']).mean()
grade_10th_reading = grade_10th_reading.rename(columns={"reading_score":"10th"})

grade_11th_reading = group_grade_reading.loc[(group_grade_reading['grade'] == '11th')].groupby(['school_name']).mean()
grade_11th_reading = grade_11th_reading.rename(columns={"reading_score":"11th"})

grade_12th_reading = group_grade_reading.loc[(group_grade_reading['grade'] == '12th')].groupby(['school_name']).mean()
grade_12th_reading = grade_12th_reading.rename(columns={"reading_score":"12th"})

```

```python
grades_9_10_reading = pd.merge(grade_9th_reading, grade_10th_reading, how="left", on=["school_name"])
grades_9_10_11_reading = pd.merge(grades_9_10_reading, grade_11th_reading, how="left", on=["school_name"])    
reading_score_by_grade = pd.merge(grades_9_10_11_reading, grade_12th_reading, how="left", on=["school_name"])
reading_score_by_grade
```
